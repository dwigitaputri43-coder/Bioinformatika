```
# --- PROGRAM ANALISIS DATA BIOLOGIS SEDERHANA ---

# 1. Variabel dan Input (Ingat: Input selalu string, harus di-convert!)
nama_sampel = input("Masukkan nama sampel: ")
urutan_dna = input("Masukkan urutan DNA (contoh: ATGCGCG): ").upper() # Memastikan kapital
konsentrasi = float(input("Masukkan konsentrasi DNA (ng/uL): "))

print(f"\n--- Memulai Analisis Sampel: {nama_sampel} ---")

# 2. Perhitungan Statistik Sederhana (GC Content)
jumlah_c = urutan_dna.count("C")
jumlah_g = urutan_dna.count("G")
total_basa = len(urutan_dna)
gc_content = ((jumlah_c + jumlah_g) / total_basa) * 100

print(f"Total Basa: {total_basa}")
print(f"GC Content: {gc_content:.2f}%")

# 3. Logika Seleksi (Struktur IF Bertingkat & Hierarki)
if konsentrasi >= 50:
    status = "Sangat Baik"
elif konsentrasi >= 20:
    status = "Cukup (Perlu Optimasi)"
else:
    status = "Rendah (Gagal)"

print(f"Status Kualitas DNA: {status}")

# 4. Loop & Kontrol Alur (Mencari Basa Non-Nukleotida)

while True:
  nama_sampel = input()
  if nama_sampel == "jamur tiram":
    break


print("\n--- Analisis Selesai ---")
```

```
print("--- Program Filter DNA (Ketik 'STOP' untuk keluar) ---")

while True:
    dna = input("\nMasukkan urutan DNA: ").upper()

    # 1. Cek pintu keluar (BREAK)
    if dna == "STOP":
        print("Program selesai. Semangat UTS-nya!")
        break

    # 2. Cek validasi (CONTINUE)
    # Kita mau skip kalau DNA-nya terlalu pendek (misal cuma 'AT')
    if len(dna) < 4:
        print("Error: DNA terlalu pendek! Coba input lagi yang bener.")
        continue # Langsung balik ke atas (baris 'while True'), baris bawah di-skip

    # 3. Proses kalau lolos sensor
    print(f"Data '{dna}' valid. Sedang memproses...")
    # Bayangkan ada proses rumit di sini...
```

```
urutan = input("Masukkan DNA: ").upper() # Biar jadi kapital semua
jumlah_bersih = 0

for basa in urutan:
    print(f"Mengecek basa: {basa}") # Buat intip prosesnya
    if basa not in "ATGC":
        continue

    jumlah_bersih = jumlah_bersih + 1
    print(f"Berhasil! Jumlah sekarang: {jumlah_bersih}")

print(f"\nTOTAL AKHIR: {jumlah_bersih}")
```
